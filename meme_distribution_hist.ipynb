```python
import matplotlib.pyplot as plt
import numpy as np

# Your code for generating random data
mean = 5
std_dev = 2
num_samples = 1000
values = np.round(np.random.normal(mean, std_dev, num_samples)).astype(int)
values = np.clip(values, 0, 10)

# Calculate the bin edges and positions for x-axis ticks
bin_edges = np.arange(-0.5, 10.5, 1)  # Adjust the range based on your data
x_tick_positions = np.arange(0, 11, 1)

# Create the histogram
plt.hist(values, bins=bin_edges, alpha=0.8)

# Add labels and title
plt.xlabel('# of pions generated')
plt.xticks(x_tick_positions, ['0 point', '1 point', '2 point', '3 point', '4 point', '5 point', '6 point', '7 point', '8 point', '9 point', '10 point'], rotation=45, ha='right')
plt.ylabel('# of memes')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your code for generating random data

values = np.round(np.random.normal(10, 2, 5000)).astype(int)
add_values = np.round(np.random.normal(0, 2, 5000)).astype(int)
values = np.append(values, add_values)
values = np.clip(values, 0, 10)

# Set the bin width to control the separation between bins
bin_width = 0.5  # Adjust this value based on your preference

# Create the histogram with the specified bin width
plt.hist(values, bins=np.arange(-0.5, 11.5, bin_width), alpha=0.8)

# Add labels and title
plt.xticks(np.arange(0, 11), ['0 point', '1 point', '2 point', '3 point', '4 point', '5 point', '6 point', '7 point', '8 point', '9 point', '10 point'], rotation=45, ha='right')
plt.ylabel('# of memes')

# Show the plot
plt.show()
```
