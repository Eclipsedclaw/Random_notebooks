# flux map (powerlaw) の入ったfitsファイルの中身を確認する

```python
import healpy as hp
import astropy.io.fits as pyfits
import numpy as np

filename = "../fitsfile/AliceSprings_Australia_2021_3_21_alt30000m_powerlaw_photon.fits"
```

```python
print(filename)
```

```python
flux_map = hp.read_map(filename, field = None)
```

```python
header = pyfits.open(filename)[1].header

n_map = header["TFIELDS"]
nside = header["NSIDE"]
map_names = []
for i in range(n_map):
    map_names.append(header["TTYPE{}".format(i+1)])

print(header)
```

```python
for i in range(n_map):
    try:
        hp.mollview(flux_map[i], title = map_names[i], unit = header["TUNIT{}".format(i+1)])
    except:
        hp.mollview(flux_map[i], title = map_names[i])
        
print(np.sum(flux_map[3]))
```
