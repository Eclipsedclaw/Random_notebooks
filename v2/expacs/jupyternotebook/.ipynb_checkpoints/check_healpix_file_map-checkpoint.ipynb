# flux map の入ったfitsファイルの中身を確認する

```python
import healpy as hp
import astropy.io.fits as pyfits
import numpy as np

filename = "../fitsfile/AliceSprings_Australia_2021_3_21_alt30000m_map_photon.fits"
```

```python
print(filename)
```

```python
flux_map = hp.read_map(filename, field = None)
```

```python
header = pyfits.open(filename)[1].header

n_map = header["TFIELDS"]
nside = header["NSIDE"]
map_names = []
for i in range(n_map):
    map_names.append(header["TTYPE{}".format(i+1)])

print(header)
```

# flux map の描画

```python
for i in range(n_map):
    hp.mollview(flux_map[i], title = map_names[i], unit = header["TUNIT{}".format(i+1)])
```

# 積分photon flux

```python
flux_sum = np.zeros(len(flux_map[0]))
e_range = np.array([header["ENE{}".format(i+1)] for i in range(len(flux_map))])
print(e_range)

for i in range(len(e_range)-1):
    flux_sum += flux_map[i] * (e_range[i+1] - e_range[i]) 

print("")
print("The integrated photon flux from {} MeV to {} MeV is".format(e_range[0], e_range[-1]), np.sum(flux_sum), "ph/cm2/s.")

hp.mollview(np.log10(flux_sum), cmap = 'plasma', unit = "ph/cm2/s/pixel")
```

# 積分エネルギーフラックス

```python
MeVtoErg = 1.602176565e-06
energyflux_sum = np.zeros(len(flux_map[0]))

for i in range(len(e_range)-1):
    energyflux_sum += flux_map[i] * (e_range[i+1] - e_range[i]) *  0.5 * (e_range[i] + e_range[i+1]) * MeVtoErg

print("")
print("The integrated energy flux from {} MeV to {} MeV is".format(e_range[0], e_range[-1]), np.sum(energyflux_sum), "erg/cm2/s.")

hp.mollview(np.log10(energyflux_sum), cmap = 'plasma', unit = "erg/cm2/s/pixel")
```
