```python
import matplotlib.pyplot as plt

Level = [x for x in range(13)]
Resistance = [2.202, 2.203, 2.087, 1.949, 1.772, 1.644, 1.511, 1.351, 1.168, 1.070, 0.905, 0.775, 0.640]

# Create the plot
plt.plot(Level, Resistance, label = 'Room Temperature water',marker='o', linestyle='-', color='b')

# Add labels and a legend
plt.xlabel('Water level from the bottom/in')
plt.ylabel('Resistance/kΩ')
plt.title('Resistance of the level strip in water under room temperature')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
